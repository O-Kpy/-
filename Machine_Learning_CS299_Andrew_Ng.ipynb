# 1강. Welcome to ML

1. 지도 학습(Supervised Learning)
 > 회귀, 분류...가 지도 학습이다.(SVM, Regression, Classifier, 인공지능 모델, 트리모델, 시계열모델... 등등)

 > X(변수 피쳐)들과 Y(레이블피쳐)가 주어지고, 주로 예측을 하는데 그 목적이 있다.

2. 비지도 학습(Unsupervised Learning)
 > 군집화, 차원축소분석(LCA빼고, PCA, ICA, SVD...), SNA(소셜 네트워크 분석)... 등등

 > X가 주어지고, 유사성, 관련성이 있는 데이터들 끼리 군집을 이룹니다. 이 군집을 통해 구조를 만들고 분류 하여 데이터의 인사이트를 얻고자 하는 것이 주 목표이다. ==> 데이터 마이닝의 성격이 강하다.

 > ex) 구글의 비슷한 뉴스를 모아서 보여주는 알고리즘

3. 강화 학습(Reinforcement Learning) 
 > 행동 집합(Action), 환경 상태 집합(State), 보상 집합(Reward)로 구성되어 있다.

 > "잘했으면 상을 주고, 못했으면 벌을 주는..."

 > 최적의 방법이 무엇인지 아무것도 모를때, 시간이 지남에 따라 나아지고 불량률도 적어진다.

 > 확률적 경사 하강법(SGD) - 비효율적이지만 글로벌 최적값으로 
 수렴해 간다.(계산이 빠르다, 데이터가 큰 상황에서 BGD보다 자주 쓰인다.)

 > BGD(Batch Gradient Descent) - 효율적이지만 계산이 느리다
